```python
import pandas as pd
import numpy as np
```

```python
#data available from Kaggle: /kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv
df = pd.read_csv('country_vaccinations.csv')

# look through the column names as we're currently only interested in the column that has vaccine names
df.columns
```

```python
# concentrating on vaccines column, we need to create a list of the same
# some countries have a combination of vaccines administered, we therefore split the entries
total_vaccines = [name.split(',') for name in df['vaccines']] 

# flattening the list of lists
flat_vaccine_list = sum(total_vaccines, [])
```

```python
# we can now create a final list with distinct vaccines and have a count therein as well
distinct_vaccine_list = []
count = 0

# we use a for loop 
for name in flat_vaccine_list:
    # removing any spaces in the entries
    name = name.strip()
    # adding distinct entries only
    if name not in distinct_vaccine_list:
        distinct_vaccine_list.append(name)
        count += 1

print(distinct_vaccine_list)
print(f'The number of administered vaccines is {count}')
        
    
```
